```python
import pandas as pd
import numpy as np
```

```python
# dataset
data = pd.read_csv("HYPO_FORMATTED.csv")

# printing the shape of the dataset
data.shape
```

```python
data.head()
```

```python
# Getting the list of transactions from the dataset
transactions = []
for i in range(0, len(data)):
    transactions.append([str(data.values[i,j]) for j in range(0, len(data.columns))])

```

```python

# Extract unique items.
flattened = [item for transaction in transactions for item in transaction]
items = list(set(flattened))
if 'nan' in items: items.remove('nan')
print(list(items))
len(items)
```

```python
def extractPairs(lst):
    if 'P' in items: items.remove('P')
    if 'N' in items: items.remove('N')
    return [[el,'P'] for el in lst]
```

```python
rules=extractPairs(items)
rules
```

```python

# Import the transaction encoder function from mlxtend
from mlxtend.preprocessing import TransactionEncoder

# Instantiate transaction encoder and identify unique items
encoder = TransactionEncoder().fit(transactions)

# One-hot encode transactions
onehot = encoder.transform(transactions)

# Convert one-hot encoded data to DataFrame
onehot = pd.DataFrame(onehot, columns = encoder.columns_).drop('nan', axis=1).drop('N',axis=1)

# Print the one-hot encoded transaction dataset
onehot.head()
```

```python
onehot.shape
```

```python
# Compute the support
support = onehot.mean()
support = pd.DataFrame(support, columns=['support']).sort_values('support',ascending=False)

# Print the support
support
```

```python
# Define a function to compute Zhang's metric
def zhang(antecedent, consequent):
    # Compute the support of each book
    supportA = antecedent.mean()
    supportC = consequent.mean()

    # Compute the support of both books
    supportAC = np.logical_and(antecedent, consequent).mean()

    # Complete the expressions for the numerator and denominator
    numerator = supportAC - supportA*supportC
    denominator = max(supportAC*(1-supportA), supportA*(supportC-supportAC))

    # Return Zhang's metric
    return numerator / denominator
```

```python
# Create rules DataFrame
rules_ = pd.DataFrame(rules, columns=['antecedents','consequents'])

# Define an empty list for metrics
zhangs, antec_supp, cons_supp, suppt = [], [], [], []

# Loop over lists in itemsets
for itemset in rules:
 
    # Extract the antecedent and consequent columns
    antecedent = onehot[itemset[0]]
    consequent = onehot['P']
    
    antecedent_support = onehot[itemset[0]].mean()
    consequent_support = onehot[itemset[1]].mean()
    support = np.logical_and(onehot[itemset[0]], onehot[itemset[1]]).mean() 
    antec_supp.append(antecedent_support)
    cons_supp.append(consequent_support)
    suppt.append(support)
    zhangs.append(zhang(antecedent, consequent))  
# Store results
rules_['antecedent support'] = antec_supp
rules_['consequent support'] = cons_supp
rules_['support'] = suppt
rules_['zhang'] = zhangs

# Print results
rules_.sort_values('zhang',ascending=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
   
data = {'Attribute':rules_['antecedents'],
        'Zhang_metric': rules_['zhang']
      }

  
df = pd.DataFrame(data,columns=['Attribute','Zhang_metric'])
df.plot(x ='Attribute', y='Zhang_metric', kind = 'bar')
plt.show()
```

```python
zhang_score=dict(zip(rules_['antecedents'], rules_['zhang']))
```

```python

def accuracy_score(error_rate):
  correct=0
  var=""
  for trans in transactions:
    association_score=0
    for i in range(0,21):
      if(trans[i]!="nan"):
    
        association_score=association_score+zhang_score[trans[i]]
    if (association_score>=(0-error_rate)): var="P"
    else: var="N"
    if(trans[21]==var):correct=correct+1
  return correct/len(transactions)
```

```python
print(accuracy_score(0))
```

```python

error_lst=[]
accuracy_lst=[]
def set_error_rate():
  error_rate=0
  prev_acc=accuracy_score(error_rate)
  incr=0.04
  
  error_lst.append(error_rate)
 
  accuracy_lst.append(prev_acc)

  while(1):
    error_change=error_rate+incr
    if(prev_acc)>accuracy_score(error_change):
      print("Association_Error:",error_rate,",Accuracy:",accuracy_score(error_rate)*100,"%")
      error_lst.append(error_change)
      accuracy_lst.append(accuracy_score(error_change))

      return error_rate
    else:
      error_rate=error_change
      prev_acc=accuracy_score(error_change)
      error_lst.append(error_rate)
      accuracy_lst.append(prev_acc)
```

```python
association_error=int(set_error_rate())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
   
data = {'Error_rate':error_lst,
        'Accuracy': accuracy_lst
      }

  
df = pd.DataFrame(data,columns=['Error_rate','Accuracy'])
df.plot(x ='Error_rate', y='Accuracy', kind = 'bar')
plt.show()
```

```python
def predict(lst):
  association_score=0
  for i in lst:
    try:
      association_score=association_score+zhang_score[i]
    except:
      association_score=association_score+0
  if (association_score>=(0-association_error)): return("PRESENCE OF HYPOTHYROIDISM")
  else: return("ABSENCE OF HYPOTHYRODISIM")

```
